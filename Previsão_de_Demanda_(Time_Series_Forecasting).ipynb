

Aqui come√ßa a prepara√ß√£o dos dados para nossa an√°lise.

Na vari√°vel `price_brl` podemos observar que cont√©m "v√≠rgula", por√©m, o Python n√£o processa valores nesse formato, sendo assim, teremos de ajustar essas informa√ß√µes.

Al√©m disso, vamos converter para float os n√∫meros monet√°rios.

J√° a vari√°vel `item_id` que est√° como n√∫mero (int), vamos alterar para string. Faremos isso para que o algoritmo que formos usar depois, n√£o tente associar a grandeza do n√∫mero de ID, e n√£o leve isso em considera√ß√£o. ###**Analisando os Dados**

Agora que temos uma ideia, um panorama desse conjunto de dados, vamos criar uma c√≥pia do conjunto e algumas novas vari√°veis que possam nos ajudar a explorar esse conjunto.

Lembrando que estamos lidando com uma **S√©rie Temporal** existem v√°rios insights que podemos tirar dos dados nesse sentido.

```python
# criando uma c√≥pia do dataset
df_clean = df.copy()
# criando mais features para a an√°lise temporal
df_clean["Year"] = df_clean.index.year
df_clean["Quarter"] = df_clean.index.quarter
df_clean["Month"] = df_clean.index.month
df_clean["Week"] = df_clean.index.week
df_clean["Weekday"] = df_clean.index.weekday
df_clean["Day"] = df_clean.index.day
df_clean["Dayofyear"] = df_clean.index.dayofyear
df_clean["Date"] = pd.DatetimeIndex(df_clean.index).date
df_clean.drop(df_clean[df_clean.Year == 2016].index, axis=0, inplace=True)

# checando as novas features
df_clean.head()
```

##**Previs√£o de Demanda | S√©ries temporais usando o Facebook Prophet**

O que s√£o s√©ries temporais?
Resumidamente, uma s√©rie temporal √© um conjunto de observa√ß√µes feitas em sequ√™ncia ao longo do tempo.

As s√©ries temporais apresentam algumas propriedades importantes:
* **Tend√™ncia**: Significa saber se uma determinada s√©rie est√° crescendo, diminuindo ou se est√° est√°vel.
* **Sazonalidade**: S√£o flutua√ß√µes peri√≥dicas, fen√¥menos que se repetem a cada per√≠odo id√™ntico do tempo. Por exemplo, a venda dos panetones que aumentam todos os anos, durante os meses de dezembro.
* **Ciclos**: √â um aumento ou redu√ß√£o da frequ√™ncia, mas sem intervalos fixos, o que difere da sazonalidade por n√£o ter um intervalo frequente.
* **Erro Aleat√≥rio**: S√£o flutua√ß√µes inexplic√°veis, resultando de fatos fortuitos e inesperados como catastrofes naturais, atentados terroristas e pandemias.

Al√©m desses componentes √© importante entender outro conceito que aparece bastante em modelos de s√©ries temporais: a estacionariedade.

Uma s√©rie estacion√°ria √© quando a m√©dia, vari√¢ncia e a estrutura de autocorrela√ß√£o se mant√©m constantes durante o tempo.

Essas informa√ß√µes foram tiradas do artigo Introdu√ß√£o a S√©ries temporais, para saber mais, [clique aqui](https://medium.com/data-hackers/series-temporais-parte-1-a0e75a512e72).

Para fazermos as previs√µes, utilizaremos o `Prophet`, que √© um software de c√≥digo aberto lan√ßado pela equipe Core Data Science do Facebook.

O Prophet √© bastante robusto, simples e f√°cil de usar. Voc√™s v√£o encontrar na internet exemplos de c√≥digo utilizando o Prophet sem nem se preoucupar se a S√©rie Temporal √© estacion√°ria, as vezes o resultado pode ser bom. Mas o ideal √© sempre verificarmos se ela √© estacion√°ria ou n√£o.

###**A s√©rie temporal √© estacion√°ria?**
Um pr√©-requisito para que possamos fazer previs√µes em cima de uma s√©rie temporal √© ela ser estacion√°ria.

Como falamos anteriormente, ser estacion√°ria significa ter a **m√©dia, vari√¢ncia e covari√¢ncia** permanecendo **constante** ao longo da dimens√£o temporal. Vamos entender visualmente o que isso significa.

**Como verificar se uma S√©rie Temporal √© estacion√°ria?**

H√° muitas maneiras se fazer essa verifica√ß√£o. Por exemplo:

* **Olhar os plots -** Em muitos casos, durante a an√°lise explorat√≥ria voc√™ j√° vai identificar visualmente se uma s√©rie temporal √© estacion√°ria ou n√£o.
* **Resumo estat√≠stico -** Voc√™ pode gerar um resumo com as principais informa√ß√µes estat√≠sticas da s√©rie temporal, e verificar se h√° algum ind√≠cio da estacionaridade.
* **Testes estat√≠sticos -** S√£o testes que possuem a capacidade de nos dizer se uma s√©rie temporal √© estacion√°ria ou n√£o, sempre se baseando em m√©todos cient√≠ficos.

Nesta projeto, realizaremos um teste estat√≠stico espec√≠fico para identificar se nossa s√©rie temporal √© estacion√°ria ou n√£o, o teste ***Augmented Dickey Fuller (ADF)***

### **Teste Augmented Dickey Fuller (ADF)**

O ADF √© um dos testes estat√≠sticos mais usados para S√©ries Temporais. Suas hip√≥teses nula e alternativa s√£o:

* A Hip√≥tese Nula ($H_0$) do teste √© que a s√©rie temporal **n√£o √© estacion√°ria**. Ou seja, possui algum tipo de depend√™ncia em rela√ß√£o ao tempo.
* A Hip√≥tese Alternativa ($H_1$) **rejeita a hip√≥tese nula**, ou seja, que a s√©rie temporal **√© estacion√°ria**.

Vamos ver como isso vai acontecer no nosso projeto espec√≠fico.


###**Preparando os Dados**

Primeiro de tudo, precisamos preparar os dados para serem utilizados com o Prophet. Ele precisa de um tipo muito espec√≠fico de dataframe para funcionar. Precisamos de:

* Data como uma coluna chamada `ds`.
* Vendas como uma coluna chamada `y`.

√â isso que vamos fazer abaixo:

```python
# preparando o dataset
df_ts=df_clean.groupby('Date',as_index=False)['sales'].sum()
df_ts['Date'] = pd.to_datetime(df_ts['Date'], format="%Y-%m-%d")
df_ts.index = pd.to_datetime(df_ts['Date'], format="%Y-%m-%d")
df_ts.drop('Date', axis=1, inplace=True)

# nomeando corretamente
df_ts = df_ts.reset_index()
df_ts.columns = ['ds', 'y']
df_ts.head()
```

Com os dados formatados, √© hora de importar o pacote e come√ßar a fazer as previs√µes!

```python
# importando o Prophet
from prophet import Prophet
```

###**Teste Augmented Dickey Fuller (ADF)**

Conforme falamos anteriormente, vamos fazer esse teste para descobrir se a nossa s√©rie √© estacion√°ria. 

> Normalmente, rejeitamos $H_0$ quando o *p-value* √© menor ou igual a um determinado n√≠vel de signific√¢ncia. Normalmente, esse n√≠vel de signific√¢ncia √© igual a 0.01 (5%), 0.05 (5%) ou at√© mesmo 0.1 (10%).

> Quanto menor o *p-value*, mais evid√™ncias contra a hip√≥tese nula voc√™ tem. Por exemplo, caso seja usado $\text{p-value} = 0.05$, rejeitamos a $H_0$ com um n√≠vel de confian√ßa de 95%. 

> Caso o valor do teste fique acima do *threshold*, falharemos em rejeitar a hip√≥tese nula. **Aten√ß√£o, isso n√£o significa que $H_0$ √© verdadeira.** Significa apenas que o experimento n√£o encontrou evid√™ncias para rejeit√°-la.

Vamos ver como a nossa s√©rie se sai.

```python
# importanto o pacote
from statsmodels.tsa.stattools import adfuller
# extraindo os valores
X = df_ts.y

# aplicando o ADF e imprimindo os resultados.
result = adfuller(X)
print('Augmented Dickey‚ÄìFuller')
print('Statistical Test: {:.4f}'.format(result[0]))
print('P Value: {:.10f}'.format(result[1]))
print('Critical Values:')
for key, value in result[4].items():
	print('\t{}: {:.4f}'.format(key, value))
```

Com um `P Value` de 0.154, poderiamos rejeitar a hip√≥tese nula com um n√≠vel de confian√ßa na casa de 85%. Vamos utilizar esse conjunto como est√° para a nossa baseline, e caso os resultados n√£o sejam bons o suficiente, vamos fazer as transforma√ß√µes.

###**Tamanho das Previs√µes**

Ao lidar com previs√µes em S√©ries Temporais precisamos entender que quanto maior o per√≠odo que estamos tentando prever, menos confian√ßa teremos nos resultados. Dessa forma, mesmo tendo anos de registros de vendas, vamos fazer as nossas previs√µes para um per√≠odo de 30 dias.

```python
# definindo o tamanho da previs√£o para trinta dias
prediction_size = 30
train_df = df_ts[:-prediction_size]
```

```python
# instantiating the model
b = Prophet()

# training the model
b.fit(train_df)

# make predictions
future = b.make_future_dataframe(periods=prediction_size);

forecast = b.predict(future)

forecast.head()
```

###**Conclus√µes - S√©rie N√£o-Estacion√°ria**
Tudo isso parece lindo e perfeito, mas como saber se nossas previs√µes est√£o boas de verdade?

Para isso utilizaremos as seguintes m√©tricas:

* Mean Absolute Percentage Error - Mostra quanto as nossas previs√µes diferem do valor real em porcentagem.
* Mean Absolute Error - Valor absoluto do nosso erro na previs√£o em rela√ß√£o √† s√©rie real.

Abaixo temos uma fun√ß√£o para calcular o MAPE e MAE da nossa s√©rie.

```python
# definindo a fun√ß√£o
def make_comparison_dataframe(historical, forecast):
  return forecast.set_index('ds')[['yhat', 'yhat_lower',
                                'yhat_upper']].join(historical.set_index('ds'))
cmp_df = make_comparison_dataframe(df_ts, forecast)

def calculate_forecast_errors(df_ts, prediction_size):

  df = df_ts.copy()
  df['e'] = df['y'] - df['yhat']
  df['p'] = 100 * df['e'] / df['y']
  predicted_part = df[-prediction_size:]

  error_mean = lambda error_name: np.mean(np.abs(predicted_part[error_name]))

  return{'MAPE': error_mean('p'), 'MAE': error_mean('e')}

# imprimindo o MAPE e MAE
for err_name, err_value in calculate_forecast_errors(cmp_df, prediction_size).items():
  print(err_name, err_value)
```

###**Transformando a s√©rie em Estacion√°ria**

> Para transformar uma TS em estacion√°ria, precisamos remover a tend√™ncia e a sazonalidade que h√° nos dados originais. Para isso, existem diversas t√©cnicas que podem ser usadas.

> Uma bom come√ßo √© aplicar o $log$ a fim de reduzirmos a magnitude dos valores da TS. Na sequ√™ncia, subtra√≠mos a m√©dia m√≥vel de um determinado per√≠odo em rela√ß√£o ao $log$ da TS.

Existe outra t√©cnica que podemos utilizar para potencializar esses resultados que √© a diferencia√ß√£o.

###**O que √© a Diferencia√ß√£o?**

Uma outra t√©cnica simples que ajuda a tornar TS mais estacion√°rias √© a diferencia√ß√£o.

Apenas calculando a diferen√ßa entre duas observa√ß√µes poss√≠veis, √© poss√≠vel incrementar ainda mais o desempenho no ADF.

$$value(t) = observation(t) - observation(t -1)$$

```python
# aplicando a diferencia√ß√£o
df_diff = df_sub.diff(1)
ma_diff = df_diff.rolling(24).mean()
std_diff = df_diff.rolling(24).std()

fig, ax = plt.subplots(figsize=(12,8))
df_diff.plot(ax=ax, legend=False)
ma_diff.plot(ax=ax, color='r', legend=False)
std_diff.plot(ax=ax, color='g', legend=False)
plt.show()

# extraindo os valores
X_diff = df_diff.y.dropna().values

# aplicando o ADF e imprimindo os resultados 
result_diff = adfuller(X_diff)
print('Augmented Dickey‚ÄìFuller')
print('Statistical Test: {:.4f}'.format(result_sub[0]))
print('P Value: {:.10f}'.format(result_sub[1]))
print('Critical Values:')
for key, value in result_sub[4].items():
	print('\t{}: {:.4f}'.format(key, value))
```
